```python
# split dependent and independent variables
x_ind=dataset.drop('Property_Sale_Price',axis=1)
x_ind
```

```python
y_dep=dataset.Property_Sale_Price
y_dep
```

```python
# perform least square model fitting to seperate none correlated variable which are under 5% signifance value
import statsmodels.api as sm
model=sm.OLS(y_dep,x_ind)
fit=model.fit()
fit.summary()
```

```python
# remove all variable with signifance value above 5%
x_ind1=x_ind.drop(["Id","Dwell_Type","LotFrontage","LotArea","YearBuilt","BsmtFinSF2","BsmtHalfBath","OpenPorchSF","EnclosedPorch","3SsnPorch","PoolArea"],axis=1)
```

```python
import statsmodels.api as sm
model=sm.OLS(y_dep,x_ind1)
fit=model.fit()
fit.summary()
```

```python
# seperate table for machine learning and evalution
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import statistics
import cmath
root_list=[]
random_state_list=[]
test_size_list=[]
for g in [.2,.3]:
    for f in range(10):
        x_train,x_test,y_train,y_test=train_test_split(x_ind1,y_dep,test_size=g,random_state=f)
    # fit the model with training set and predict the value
        LR_model=LinearRegression()
        LR_model.fit(x_train,y_train)
        pred=LR_model.predict(x_test)
    #calculate the residual 
        df=pd.DataFrame({'actual':y_test,'pre':pred})
    # To find the error  in the predicted and actual value
        residual=pred-y_test
    # find the root mean square of residual
        mean=statistics.mean(residual)
        root=cmath.sqrt(mean)
        root_list.append(root)
        random_state_list.append(f)
        test_size_list.append(g)
finallist=pd.DataFrame({'test_size':test_size_list,'random_state':random_state_list,'root':root_list})
finallist
```
